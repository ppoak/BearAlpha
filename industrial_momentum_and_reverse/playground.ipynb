```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
industry = list(map(lambda x: x.rstrip('.csv'), os.listdir('data/processed/')))
```

```python
dataset = [pd.read_csv(f"data/processed/{ind}.csv",
    index_col=0, parse_dates=['日期']) for ind in industry]
```

```python
close_price = [data['收盘价(元)'].resample('d').last().dropna() for data in dataset]
open_price = [data['开盘价(元)'].resample('d').first().dropna() for data in dataset]
```

```python
overnight_ret = [np.log(op / cp.shift(1)).dropna() for (op, cp) in zip(open_price, close_price)]
intraday_ret = [np.log(cp / op).dropna() for (op, cp) in zip(open_price, close_price)]
raw_ret = [np.log(cp / cp.shift(1)).dropna() for cp in close_price]
overnight_ret = pd.concat(overnight_ret, axis=1)
overnight_ret.columns = industry
intraday_ret = pd.concat(intraday_ret, axis=1)
intraday_ret.columns = industry
raw_ret = pd.concat(raw_ret, axis=1)
raw_ret.columns = industry
```

```python
overnight_ret_m = overnight_ret.resample('m').sum()
intraday_ret_m = intraday_ret.resample('m').sum()
raw_ret_m = raw_ret.resample('m').sum()
```

```python
forward_ret_m = raw_ret_m.rolling(window=1).sum().shift(-1).dropna()
forward_ret_m
```

```python
overnight_corr = np.mean(forward_ret_m.columns.map(lambda x: forward_ret_m[x].corr(overnight_ret_m[x])))
intraday_corr = np.mean(forward_ret_m.columns.map(lambda x: forward_ret_m[x].corr(intraday_ret_m[x])))
```

```python
overnight_arg_sort = np.argsort(overnight_ret_m)
overnight_portfolio = pd.DataFrame(index=overnight_arg_sort.index, columns=overnight_arg_sort.columns)
for idx in overnight_portfolio.index:
    overnight_portfolio.loc[idx, overnight_portfolio.columns[overnight_arg_sort.loc[idx].iloc[:5]]] = 1 / 5
overnight_portfolio

intraday_arg_sort = np.argsort(intraday_ret_m)
intraday_portfolio = pd.DataFrame(index=intraday_arg_sort.index, columns=intraday_arg_sort.columns)
for idx in intraday_portfolio.index:
    intraday_portfolio.loc[idx, intraday_portfolio.columns[intraday_arg_sort.loc[idx].iloc[:5]]] = 1 / 5
intraday_portfolio
```

```python
intraday_portfolio_ret = ((np.exp(forward_ret_m) - 1) * intraday_portfolio).sum(axis=1)
intraday_portfolio_ret = np.log(intraday_portfolio_ret + 1)
overnight_portfolio_ret = ((np.exp(forward_ret_m) - 1) * overnight_portfolio).sum(axis=1)
overnight_portfolio_ret = np.log(overnight_portfolio_ret + 1)
```

```python
intraday_portfolio_ret.mean() * 100, overnight_portfolio_ret.mean() * 100
```

```python
intraday_portfolio_ret.mean() / intraday_portfolio_ret.std() * np.sqrt(len(intraday_portfolio_ret)), \
overnight_portfolio_ret.mean() / overnight_portfolio_ret.std() * np.sqrt(len(overnight_portfolio_ret))
```

```python
dataset = [pd.read_csv(f'data/daily.nosync/{code}.csv', index_col=0, parse_dates=['TRADE_DT'])
    for code in industry]
```

```python
raw_ret = [np.log(data['S_DQ_CLOSE'] / data['S_DQ_CLOSE'].shift(1)).dropna() for data in dataset]
overnight_ret = [np.log(data['S_DQ_OPEN'] / data['S_DQ_CLOSE'].shift(1)).dropna() for data in dataset]
intraday_ret = [np.log(data['S_DQ_CLOSE'] / data['S_DQ_OPEN']).dropna() for data in dataset]

raw_ret = pd.concat(raw_ret, axis=1)
raw_ret.columns = industry
overnight_ret = pd.concat(overnight_ret, axis=1)
overnight_ret.columns = industry
intraday_ret = pd.concat(intraday_ret, axis=1)
intraday_ret.columns = industry
```

```python
raw_ret_m = raw_ret.resample('m').sum()
overnight_ret_m = overnight_ret.resample('m').sum()
intraday_ret_m = intraday_ret.resample('m').sum()
```

```python
raw_argsort = np.argsort(raw_ret_m)
raw_portfolio = pd.DataFrame(index=raw_argsort.index, columns=raw_argsort.columns)
for idx in raw_argsort.index:
    raw_portfolio.loc[idx, raw_argsort.columns[raw_argsort.loc[idx].iloc[:5]]] = 1 / 5

overnight_argsort = np.argsort(overnight_ret_m)
overnight_portfolio = pd.DataFrame(index=overnight_argsort.index, columns=overnight_argsort.columns)
for idx in overnight_argsort.index:
    overnight_portfolio.loc[idx, overnight_argsort.columns[overnight_argsort.loc[idx].iloc[:5]]] = 1 / 5

intraday_argsort = np.argsort(intraday_ret_m)
intraday_portfolio = pd.DataFrame(index=intraday_argsort.index, columns=intraday_argsort.columns)
for idx in intraday_argsort.index:
    intraday_portfolio.loc[idx, intraday_argsort.columns[intraday_argsort.loc[idx].iloc[:5]]] = 1 / 5
```

```python
forward_ret_m = raw_ret_m.rolling(window=1).sum().shift(-1).dropna()
```

```python
raw_portfolio_ret = (forward_ret_m * raw_portfolio).sum(axis=1)
overnight_portfolio_ret = (forward_ret_m * overnight_portfolio).sum(axis=1)
intraday_portfolio_ret = (forward_ret_m * intraday_portfolio).sum(axis=1)
```

```python
raw_portfolio_ret.mean(), overnight_portfolio_ret.mean(), intraday_portfolio_ret.mean()
```
